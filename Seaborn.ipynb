```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.scatterplot( x=iris.sepal_length,y=iris.sepal_width)    #to plot scatter graph to see relation btw sepal_length and sepal_width
```

```python
sns.scatterplot( x=iris.sepal_length,y=iris.petal_length)
```

```python
sns.scatterplot( x=iris.petal_width,y=iris.petal_length)
```

```python
sns.displot(iris['sepal_length'])       #distribution plot is equal to hisrtogram and this represent data in block pattern as we see
# it defines on the basis of frequency of data availabe in range
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot( x=tips.tip,y=tips.total_bill)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()     # it gives  count of smoker and non smoker fromm the data
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip ,data=tips, hue ='smoker')   #hue do color coding according to  smoker yes or no and relplit drwa relation graph  on the basis of value in hue
```

```python
tips.head()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip ,data=tips, hue ='size')     #it also shows us different no. of category present in particular sectio   like in this data 6type of size are present
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip ,data=tips, style ='smoker')
```

```python
#style represent different grroup in a diff style like dot represent yes and cross represent no
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.jointplot(x=tips.total_bill ,y=tips.tip)     #to shows histogram and scatter plot together of provided x and y value
```

```python
sns.pairplot(tips) # used to plot graph or column with each other column  shows both histogram and scatter both
```

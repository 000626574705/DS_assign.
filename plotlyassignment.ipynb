```python
import plotly.graph_objects as pl
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
fig =pl.Figure()
fig.add_trace(pl.Scatter(x=titanic.age,y=titanic.fare,mode='markers'))
```

```python
#Q2 solution-
tips = sns.load_dataset('tips')
import plotly.express as px
fig = px.box(tips,y='total_bill')
fig.show()
```

```python
#Q3 solution-
import plotly.express as px

fig = px.histogram(tips, x="sex", y="total_bill", color="day", barmode="group",
                   facet_col="smoker", pattern_shape="day")
fig.show()
```

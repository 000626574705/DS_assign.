solution-Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive statistical graphics. Here are five commonly used plots in Seaborn and their uses:

*   1.Scatter plot:
Use: Visualize the relationship between two numerical variables.
*   2.Bar Plot:
Use: Displaying categorical data using rectangular bars


* 3.  Histogram:

Use: Visualize the distribution of a single numerical variable.

* 4.  Box Plot (Box-and-Whisker Plot):

Use: Visualize the distribution of numerical data and identify outliers.



* 5.  Heatmap:

Use: Display the correlation between multiple variables using color intensity.








```python
#Q2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
fmri = sns.load_dataset('fmri')

df=pd.DataFrame(fmri)
sns.lineplot(x="timepoint", y="signal" , hue='region',data=df)
```

```python
#Q6
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'flights' dataset from Seaborn
flights = sns.load_dataset('flights')

# Pivot the data to create a 2D matrix for the heatmap
flights_pivot = flights.pivot('month', 'year', 'passengers')

# Create a heatmap using Seaborn
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
# Show the plot
plt.show()
```

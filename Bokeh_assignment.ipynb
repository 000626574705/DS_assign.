```python
#Q2
#In Bokeh, glyphs are the basic visual building blocks used to represent data on a plot. Glyphs are graphical shapes or markers that can be used to display data points, lines, bars, areas, and more. Each type of plot (e.g., scatter plot, line plot, bar plot) in Bokeh is created by adding specific glyphs to the plot.

#To add glyphs to a Bokeh plot, WE'll need to use the glyph methods available in the Bokeh figure object. Some common glyph methods include line(), scatter(), circle(), square(), rect(), and vbar().

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.sampledata.iris import flowers

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 8]
category = ['A', 'B', 'C', 'D', 'E']

# Create a figure
p = figure(title='Bokeh Plot with Multiple Glyphs', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add glyphs to the figure
# Line glyph
p.line(x, y, line_color='blue', line_width=2)

# Circle glyph for scatter points
p.circle(x, y, size=10, color='red', alpha=0.8)

# Text glyph for labels
p.text(x, y, text=category, text_color='black', text_align='center', text_font_size='12pt')

# Rect glyph for bars
p.rect(x=[6, 7, 8, 9, 10], y=[2, 5, 3, 8, 6], width=0.5, height=[0.8, 1.2, 0.5, 1.0, 0.7], color='green', alpha=0.6)

# Show the plot
output_notebook()
show(p)

```

```python
#Q3 solution-
#  Bokeh provides a wide range of options to customize the appearance of a plot, including the axes, title, and legend. You can modify various attributes of the plot and its components to achieve the desired look. Here's a breakdown of how to customize different aspects of a Bokeh plot:

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare your data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 8]

# Create a figure
p = figure(title='Customized Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add data to the figure
p.line(x, y, line_color='blue', line_width=2, legend_label='Line Plot')
p.circle(x, y, size=10, color='red', alpha=0.8, legend_label='Scatter Points')

# Customize axes
p.xaxis.axis_label_text_color = 'green'
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_color = 'purple'
p.yaxis.axis_label_text_font_size = '14pt'

# Customize title
p.title.text_color = 'orange'
p.title.text_font_size = '18pt'
p.title.align = 'center'

# Customize legend
p.legend.location = 'top_left'
p.legend.label_text_color = 'navy'
p.legend.label_text_font_size = '12pt'
p.legend.background_fill_alpha = 0.6
p.legend.border_line_color = 'black'

# Show the plot
output_notebook()
show(p)
```
